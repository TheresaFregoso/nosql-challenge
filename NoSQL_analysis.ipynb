### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
print("Number of documents in result (hygiene = 20):", establishments.count_documents(query))
# Display the first document in the results using pprint
establishments_docs = db.establishments.find_one(query)

# Print one document
pprint(establishments_docs)
```

```python
# Convert the result to a Pandas DataFrame
# Fetch all documents matching the query
establishments_docs = list(establishments.find(query))

# Convert the result to a Pandas DataFrame
bad_hygiene = pd.DataFrame(establishments_docs)

# Display the number of rows in the DataFrame
print("Number of rows in result: ", len(bad_hygiene))


# Display the first document in the results using pprint
pprint(bad_hygiene[0:10])
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    'LocalAuthorityName': {'$regex':'London'},
    'RatingValue': {'$gte': 4}
    }

# Use count_documents to display the number of documents in the result
print("Number of documents in result (Ratings Value > or = to 4):", establishments.count_documents(query))
# Display the first document in the results using pprint
establishments_docs = db.establishments.find_one(query)

# Print one document
pprint(establishments_docs)
```

```python
# Convert the result to a Pandas DataFrame
establishments_docs = list(establishments.find(query))
good_rating = pd.DataFrame(establishments_docs)

# Display the number of rows in the DataFrame
print("Number of rows in result: ", len(good_rating))
# Display the first 10 rows of the DataFrame
pprint(good_rating[0:10])

```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

# Ensure the collection has a geospatial index on the geocode field
db.establishments.create_index([("geocode", GEOSPHERE)])

# Define the query to find establishments with a RatingValue of 5
rating_query = {'RatingValue': 5}

# Define the location of "Penang Flavours"
penang_flavours_location = [0.08384000, 51.49014200] 

# Define the geospatial query and sorting criteria
near_query = {
    'RatingValue': 5,
    'geocode': {
        '$near': {
            '$geometry': {
                'type': "Point",
                'coordinates': penang_flavours_location
            },
            '$maxDistance': 1000  # 1000 meters (1km) radius
        }
    }
}

# Fetch the top 5 establishments matching the criteria, sorted by Hygiene score
establishments_docs = list(establishments.find(near_query).sort('scores.Hygiene', 1).limit(5))

# Print the results
for establishment in establishments_docs:
    pprint(establishment)
```
